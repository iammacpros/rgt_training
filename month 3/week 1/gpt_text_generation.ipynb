## Importing libraries

```python
# from transformers import GPT2LMHeadModel, GPT2Tokenizer
from transformers import AutoModelForCausalLM, AutoTokenizer
```

## Loading the pre-trained model

```python
model_name = "google/gemma-3-270m-it"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
print(f"\n Model loaded successfully!")
```

## Creating a text generating function

```python
def generate_texts(prompt, max_length=200, num_return_sequences=1, temperature=0.8, top_k=50, top_p=0.95):
    
    # tokenize input prompt
    inputs = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True)
    # Generate text
    outputs = model.generate(**inputs, 
                            max_new_tokens=150, 
                            temperature=temperature, 
                            top_k=top_k, 
                            top_p=top_p, 
                            do_sample=True)

    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    # Decode all generated sequences
    generated_message = [tokenizer.decode(outputs[0], skip_special_tokens=True)]
    return generated_message
    
print("Text generation function created!")
```

 ### Paramters in function explained

    Parameters:
        prompt: Input text to continue

        max_length: Maximum length of generated text (in tokens)

        num_return_sequences: Number of different outputs to generate

        temperature: Controls randomness (higher = more creative)

        top_k: Limits sampling to top k most likely tokens

        top_p: Nucleus sampling threshold
    

    Returns:
        List of generated text sequences
  

## Generate text with different prompts

### Prompt 1

```python
prompt1 = "Once upon a time in a distant galaxy,"

print("=" * 70)
print("PROMPT 1: Creative Story Beginning")
print("=" * 70)
print(f"Input: {prompt1}\n")

result1 = generate_texts(
    prompt1,
    max_length=100,
    num_return_sequences=2,
    temperature=0.8,
    top_k=50,
    top_p=0.95
)

for i, output in enumerate(result1, 1):
    print(f"\nGeneration {i}:")
    print("-" * 70)
    print(output)
    print()
```

### Prompt 2

```python
prompt2 = "Artificial intelligence is revolutionizing healthcare by"

print("=" * 70)
print("PROMPT 2: Technical Explanation")
print("=" * 70)
print(f"Input: {prompt2}\n")

result2 = generate_texts(
    prompt2,
    max_length=120,
    num_return_sequences=2,
    temperature=0.5,  # Lower temperature for more focused output
    top_k=40,
    top_p=0.9
)

for i, output in enumerate(result2, 1):
    print(f"\nGeneration {i}:")
    print("-" * 70)
    print(output)
    print()
```

### Prompt 3

```python
prompt3 = "Q: What is the meaning of life?\nA:"

print("=" * 70)
print("PROMPT 3: Conversational Response")
print("=" * 70)
print(f"Input: {prompt3}\n")

result3 = generate_texts(
    prompt3,
    max_length=80,
    num_return_sequences=3,
    temperature=0.9,  # Higher temperature for more diverse answers
    top_k=50,
    top_p=0.95
)

for i, output in enumerate(result3, 1):
    print(f"\nGeneration {i}:")
    print("-" * 70)
    print(output)
    print()
```
